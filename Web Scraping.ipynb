```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.bbc.com/news"
```

```python
Page = requests.get(URL)
```

```python
Articles = BeautifulSoup(Page.text, 'html')
print(Articles)
```

```python
print(Articles.prettify())
```

```python
titles = Articles.find_all('h2', class_="sc-8ea7699c-3 igpHQe")
print(titles)
```

```python
Article_title = [title.text.strip() for title in titles]
print(Article_title)
```

```python
pd.DataFrame(columns = Article_title)
```

```python
df = pd.DataFrame(columns = Article_title)
```
