```python
import pandas as pd
import openpyxl 
```

```python
df_csv = pd.read_csv("C:\\Users\\Public\\Downloads\\order_details.csv")
df_csv.head()
```

```python
df = pd.read_csv("C:\\Users\\Public\\Downloads\\Orders.csv")
df.head()
```

```python
#Handling Missing Values
# Filling missing values with a specific value
df['ShipRegion'] = df['ShipRegion'].fillna('SP')
print(df['ShipRegion'])
```

```python
#removing whitespaces
df['ShipName']= df['ShipName'].str.strip()
```

```python
#Renaming column name
df_csv.rename(columns={'orderID': 'OrderID'}, inplace=True)
print(df_csv)
```

```python
#converting the orderID to a string data type
df_csv = df_csv['OrderID'].astype('str')
df_csv
```

```python
#converting the orderID to a string data type
df = df['OrderID'].astype('str')
df
```

```python
# Dropping rows with missing values
df = df.dropna()
```

```python
#removing duplicates
df = df.drop_duplicates()
```

```python
#Merging DataFrames
merged_df = pd.merge(df, df_csv, on='OrderID', how='inner')
```
